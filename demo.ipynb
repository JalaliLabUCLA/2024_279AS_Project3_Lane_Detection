```python
def show_all_images(dir):
  images = []
  for img_path in sorted(glob.glob(dir+'/*')):
      images.append(mpimg.imread(img_path))
  plt.figure(figsize=(20,10))
  columns = 5
  for i, image in enumerate(images):
      plt.subplot(int(len(images) / columns + 1), columns, i + 1)
      plt.imshow(image)
      plt.xticks([])
      plt.yticks([])
```
